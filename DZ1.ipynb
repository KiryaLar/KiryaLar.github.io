```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def p_r(r, p_k=200.987, r_k=500, r_s=0.1, Q_m3_sut=100, mu_sPz=0.3, B=1, k_mD=50, h_m=20):
  return p_k - (18.41*Q_m3_sut*mu_sPz*B)*np.log(r_k/r_r)/(k_mD*h_m)
```

```python
r_r=np.linspace(0.1, 500,250)
```

```python
p=p_r(r=r_r)
```

```python
fig1=plt.plot(r_r,p)
```

```python
fig1=plt.plot(-r_r,p)
```

```python
plt.xlabel('Расстояние, м')
```

```python
plt.ylabel('Давление, бар')
```

```python
plt.title('Воронка депрессии')
```
